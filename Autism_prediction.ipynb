```python
#replacing Outliers to median values
def replace_outliers(df,column):
  q1 = df[column].quantile(0.25)
  q2 = df[column].quantile(0.75)
  IQR = q2 - q1
  lower_bound = q1 - (1.5*IQR)
  upper_bound = q2 + (IQR*1.5)
  median = df[column].median()
  df[column] = np.where(df[column] < lower_bound,median,df[column])
  df[column] = np.where(df[column] > upper_bound,median,df[column])
  return df
```

```python
features = df.drop('Class/ASD',axis=1)
target = df['Class/ASD']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=0.3,random_state=90141)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=1001)
x_train_sampled,y_train_sampled = smote.fit_resample(x_train,y_train)
```

```python
print(y_train.shape)
print(y_train_sampled.shape)
```

```python
print(y_train.value_counts())
print(y_train_sampled.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
#Inorder to run the code perfectly this version of libraries are very very important other wise we will get into an error.
#!pip install scikit-learn==1.3.2 imbalanced-learn==0.13.0
```

```python
cv_scores = {}
models = {
    'RandomForestClassifier':RandomForestClassifier(random_state=100),
    'DecisionTreeClassifier':DecisionTreeClassifier(random_state=100),
    'XGBClassifier':XGBClassifier(random_state=100)
}

for model_name,model in models.items():
  scores = cross_val_score(model,x_train_sampled,y_train_sampled,cv=5,scoring='accuracy')
  cv_scores[model_name]=np.mean(scores)
```

```python
cv_scores
```

```python
from sklearn.model_selection import RandomizedSearchCV

models = {
    'random_forest':{
        'model':RandomForestClassifier(),
        'param_grid':{
    'n_estimators':[100,200,300,400,500],
    'max_depth':[10,20,30,40,50],
    'min_samples_split':[2,3,4,5,10],
    'min_samples_leaf':[1,2,3,4,5],
    'bootstrap':[True,False]
      }
    },

    'decision_tree':{
        'model':DecisionTreeClassifier(),
        'param_grid':{
    'criterion':['gini','entropy'],
    'max_depth':[None,5,10,15,20,30,40,50],
    'min_samples_split':[2,3,4,5,10],
    'min_samples_leaf':[1,2,3,4,5]
        }
    },

    'xgb':{
        'model':XGBClassifier(),
        'param_grid':{
    'n_estimators':[100,200,300,400,500],
    'max_depth':[10,20,30,40,50],
    'learning_rate':[0.01,0.1,0.2,0.3,0.4],
    'subsample':[0.5,0.6,0.7,0.8,0.9,1.0],
    'colsample_bytree':[0.5,0.6,0.7]
        }
    }
}
```

```python
scores = {}
model_estimator = {}
for model_name,model_params in models.items():
  model = RandomizedSearchCV(model_params['model'],model_params['param_grid'],cv=5,scoring='accuracy')
  model.fit(x_train_sampled,y_train_sampled)
  scores[model_name] = model.best_score_
  model_estimator[model_name] = model.best_estimator_
```

```python
print(scores)
```

```python
print(model_estimator)
```

```python
with open('autism_model.pkl','wb') as file:
  pickle.dump(model_estimator,file)
```

```python
y_test_pred = {}

for model_name,model in model_estimator.items():
  y_test_pred[model_name] = model.predict(x_test)
```

```python
y_test_pred
```

```python
for model_name,y_pred in y_test_pred.items():
  print(f"{model_name} : {accuracy_score(y_test,y_pred)}")
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

for model_name,y_pred in y_test_pred.items():
  print(f"{model_name} : {confusion_matrix(y_test,y_pred)}")
  print('.'*70)

print('-'*100)

for model_name,y_pred in y_test_pred.items():
  print(f"{model_name} : {classification_report(y_test,y_pred)}")
  print('.'*70)
```
